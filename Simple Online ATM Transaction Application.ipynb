```python
import psycopg2
import pandas as pd
import random

class bank:
    
    def __init__(self):
        pass
    
    #----------------------------------------------------------------#
    def check_balance(self,name,ac_no,dbtCardNo,cvv,pwd):

        conn=psycopg2.connect(user='postgres',password='*****',
                              host='localhost',
                              port=5432,database='banking')
        cursor=conn.cursor()

        query=f"SELECT t1.ac_no,t2.debt_card_no,\
                    t1.name,t1.balance\
                    FROM customer t1\
                    INNER JOIN debitcard t2\
                    ON t1.ac_no=t2.acc_no\
                    WHERE t1.name='{name}'\
                    AND t2.debt_card_no='{dbtCardNo}'\
                    AND t1.ac_no={ac_no}\
                    AND t2.cvv='{cvv}'\
                    AND t1.password={pwd}"

        cursor.execute(query)
        balance=cursor.fetchall()
        bal_info=pd.DataFrame(balance,
                      columns=[i[0] for i in cursor.description])
        conn.commit()
        cursor.close()
        conn.close()
        return bal_info
    #-----------------------------------------------------------------#
    
    def update_balance(self,wdrn_amount,name,ac_no,dbtCardNo,cvv,pwd):
        
        conn=psycopg2.connect(user='postgres',password='*****',
                          host='localhost',
                          port=5432,database='banking')
        cursor=conn.cursor()

        query1=f"SELECT t1.balance FROM customer t1\
                INNER JOIN debitcard t2\
                ON t1.ac_no = t2.acc_no\
                AND t1.name='{name}'\
                AND t1.ac_no={ac_no}\
                AND t2.debt_card_no='{dbtCardNo}'\
                AND t2.cvv='{cvv}'\
                AND t1.password={pwd}"              

        query2=f"UPDATE customer AS c\
                SET balance = balance-{wdrn_amount}\
                FROM debitcard AS d\
                WHERE c.ac_no = d.acc_no\
                AND c.name='{name}'\
                AND c.ac_no={ac_no}\
                AND d.debt_card_no='{dbtCardNo}'\
                AND d.cvv='{cvv}'\
                AND password={pwd}"

        cursor.execute(query1)
        balance=cursor.fetchone()
        if balance[0]>wdrn_amount and wdrn_amount>0:
            cursor.execute(query2)
            conn.commit()
            cursor.close()
            conn.close()
            x=f'\n{wdrn_amount}INR sucessfully withdrawn....'
            return print(x)
        else:
            conn.commit()
            cursor.close()
            conn.close()
            return print('\n!!!! Insufficient Balance !!!!')
            
        
     #-----------------------------------------------------------------#
    def transac_id_maker(self):
        
        conn=psycopg2.connect(user='postgres',password='*****',
                          host='localhost',
                          port=5432,database='banking')
        cursor=conn.cursor()
        query="SELECT transaction_id FROM customer_audit"
        cursor.execute(query)
        trans_id=cursor.fetchall()
        trans_id_df=pd.DataFrame(trans_id,
                    columns=[i[0] for i  in cursor.description])

        newTransId=str(random.randint(10000000,99999999))

        i=0
        while newTransId in trans_id_df.values:
            newTransId=str(random.randint(10000000,99999999))
            i=i+1

        conn.commit()
        cursor.close()
        conn.close()  
        return newTransId
    #-----------------------------------------------------------------#
    def audit_trigger(self,transID,amount):
        conn=psycopg2.connect(user='postgres',password='*****',
                          host='localhost',
                          port=5432,database='banking')
        cursor=conn.cursor()

        query1="DROP TRIGGER IF EXISTS bank_audit_trg\
                ON customer;"

        query2=f"CREATE OR REPLACE FUNCTION auditlog()\
                RETURNS TRIGGER AS $bank_audit_trg$\
                BEGIN\
                INSERT INTO customer_audit\
                (withdrawn_time,acc_no,transaction_id,withdrawn_amount,balance)\
                VALUES(current_timestamp,NEW.ac_no,{transID},{amount},NEW.balance);\
                RETURN NULL;\
                END;\
                $bank_audit_trg$ LANGUAGE plpgsql;"

        query3="CREATE TRIGGER bank_audit_trg\
                AFTER UPDATE ON customer\
                FOR EACH ROW EXECUTE PROCEDURE auditlog();"

        cursor.execute(query1)
        cursor.execute(query2)
        cursor.execute(query3)

        conn.commit()
        cursor.close()
        conn.close()
        return 
    #-----------------------------------------------------------------#
    def generate_auditLog(self):
        conn=psycopg2.connect(user='postgres',password='*****',
                          host='localhost',
                          port=5432,database='banking')
        cursor=conn.cursor()

        query="SELECT * FROM customer_audit"

        cursor.execute(query)

        auditLog=cursor.fetchall()
        auditLog=pd.DataFrame(auditLog,
                    columns=[i[0] for i  in cursor.description])

        conn.commit()
        cursor.close()
        conn.close()

        return auditLog
    #-----------------------------------------------------------------#

class customer:
    
    def cust_input(self):
        conn=psycopg2.connect(user='postgres',password='*****',
                          host='localhost',
                          port=5432,database='banking')
        cursor=conn.cursor()
        #-----------------------------------------------------------------#
        while True:
            name  = input('Enter your name:')
            query1=f"SELECT name FROM customer"
            cursor.execute(query1)
            Name=cursor.fetchall()
            name_df=pd.DataFrame(Name,columns=[i[0] for i in cursor.description])
            if name in name_df.name.values:
                break
            else:
                print('Enter valid name')
        #-----------------------------------------------------------------#
        while True:
            ac_no = int(input('Enter your account no:'))
            query2=f"SELECT ac_no FROM customer\
                     WHERE name='{name}'"               
            cursor.execute(query2)
            accountNo=cursor.fetchone()
            if ac_no==accountNo[0]:
                break
            else: 
                print('Enter valid account number')
        #-----------------------------------------------------------------#
        while True:
            dbtCardNo=input('Enter Debit card number:')
            query3=f"SELECT t2.debt_card_no FROM customer t1\
                     INNER JOIN debitcard t2\
                     ON t1.ac_no = t2.acc_no\
                     AND t1.name='{name}'\
                     AND t1.ac_no={ac_no}"
            cursor.execute(query3)
            debitNo=cursor.fetchone()
            if dbtCardNo==debitNo[0]:
                break
            else:
                print('Enter valid debit card number')
        #-----------------------------------------------------------------#
        while True:
            cvv=input('Enter cvv number:')
            query4=f"SELECT t2.cvv FROM customer t1\
                     INNER JOIN debitcard t2\
                     ON t1.ac_no = t2.acc_no\
                     AND t1.name='{name}'\
                     AND t1.ac_no={ac_no}\
                     AND t2.debt_card_no='{dbtCardNo}'"
            cursor.execute(query4)
            CVV=cursor.fetchone()
            if cvv==CVV[0]:
                break
            else:
                print('Enter valid cvv number')
        #-----------------------------------------------------------------#
        while True:
            pwd = int(input('Enter your password:'))
            query5=f"SELECT t1.password FROM customer t1\
                     INNER JOIN debitcard t2\
                     ON t1.ac_no = t2.acc_no\
                     AND t1.name='{name}'\
                     AND t1.ac_no={ac_no}\
                     AND t2.debt_card_no='{dbtCardNo}'\
                     AND t2.cvv='{cvv}'"
            cursor.execute(query5)
            PWD=cursor.fetchone()
            if pwd==PWD[0]:
                break
            else:
                print('Enter valid password')
        #-----------------------------------------------------------------#        
        conn.commit()
        cursor.close()
        conn.close()
        
        return name,ac_no,dbtCardNo,cvv,pwd
                        

Bank_A=bank()
customer_1=customer()    
```
